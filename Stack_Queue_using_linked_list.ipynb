```python
class Node:
    def __init__(self,data):
        self.data=data 
        self.next=None
        

class SinglyLinkedList:
    def __init__(self):
        self.head=None
        self.size=0

    def appendleft(self,n_data):
        new_node=Node(n_data)
        if self.head==None:
            self.head=new_node

        else:
            new_node.next=self.head
            self.head=new_node
        self.size+=1

    def appendRight(self,n_data):
        new_node=Node(n_data)
        if self.head==None:
            self.head=new_node

        else:
            curr=self.head
            while curr.next!=None:
                curr=curr.next
            curr.next=new_node
        self.size+=1

    def popleft(self):
        if self.head==None:
            print("Empty list")

        else:
            self.head=self.head.next
        self.size-=1

    def popright(self):
        if self.head==None:
            print("Empty list")

        else:
            curr=self.head
            while curr.next.next!=None:
                curr=curr.next

            curr.next=None
        self.size-=1

    def __str__(self):
        st=""
        curr=self.head

        while curr!=None:
            st+=str(curr.data)+" "

            curr=curr.next

        return st



sll=SinglyLinkedList()
sll.appendleft(5)
sll.appendRight(6)
sll.appendleft(4)
sll.appendRight(7)
sll.appendleft(3)
sll.appendRight(8)
print(sll)
print(sll.size)
```

```python
class Stack:
    def __init__(self):
        self.stk_list=SinglyLinkedList()

    def push(self,data):
        self.stk_list.appendleft(data)

    def pop(self):
        self.stk_list.popleft()

    def peek(self):
        if self.isEmpty()==False:

            return self.stk_list.head.data

    def isEmpty(self):
        if self.stk_list.head==None:
            return True
        else:
            return False
        
    def size(self):
        return self.stk_list.size
    
    def print_stk(self):
        if self.isEmpty()==True:
            print("Empty stack")
        else:
            print(self.stk_list)

        
stk1=Stack()
print(stk1.isEmpty())
print(stk1.size())
stk1.push(1)
stk1.push(2)
stk1.push(3)
stk1.push(4)
stk1.push(5)
stk1.print_stk()
print(stk1.size())
stk1.pop()
print(stk1.size())
print(stk1.peek())
```
